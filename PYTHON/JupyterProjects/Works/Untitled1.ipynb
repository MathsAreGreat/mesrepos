```python
import json
import pandas as pd
from pathlib import Path
```

# Step 1: Load the Excel file

```python
DATA_FILE = Path("/home/mohamed/Documents/Projects/Python/flask_students/students.json")
with DATA_FILE.open("r") as f:
    data = json.load(f)
df = pd.read_excel("/home/mohamed/Documents/Projects/Python/MassarVerified/Files/export_notesCC_1APIC-3_0019.xlsx") 
```

```python
fd = df.loc[16:, "C":"K"]
fd = fd[["C", "D", "F","G","I","K"]]
js = fd.to_dict()
for k,v in js["C"].items():
    infos = data[v]
    infos.update(
            {
            "ID_number": v,
            "full_name": " ".join(e.title() for e in js["D"][k].split(" ") if e),
            "birthdate": js["F"][k],
            "mark1": float(js["G"][k]),
            "mark2": float(js["I"][k]),
            "mark3": float(js["K"][k]),
            "average": 0,
            "classe": "1APIC3"
        }
    )
    data[v] = infos
```

```python
with DATA_FILE.open("w") as f:
    json.dump(data, f, indent=4)
```
